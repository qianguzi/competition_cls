```python
import os
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
sns.set()
result_dir = '/home/jun/mynb/lcz/result/protein'
```

```python
result0 = pd.read_csv(os.path.join(result_dir, 'submission-20.csv'))
result0.head()
```

```python
label_names = {
    0:  "Nucleoplasm",  
    1:  "Nuclear_membrane",   
    2:  "Nucleoli",   
    3:  "Nucleoli_fibrillar_center",   
    4:  "Nuclear_speckles",
    5:  "Nuclear_bodies",   
    6:  "Endoplasmic_reticulum",   
    7:  "Golgi_apparatus",   
    8:  "Peroxisomes",   
    9:  "Endosomes",   
    10:  "Lysosomes",   
    11:  "Intermediate_filaments",   
    12:  "Actin_filaments",   
    13:  "Focal_adhesion_sites",   
    14:  "Microtubules",   
    15:  "Microtubule_ends",   
    16:  "Cytokinetic_bridge",   
    17:  "Mitotic_spindle",   
    18:  "Microtubule_organizing_center",   
    19:  "Centrosome",   
    20:  "Lipid_droplets",   
    21:  "Plasma_membrane",   
    22:  "Cell_junctions",   
    23:  "Mitochondria",   
    24:  "Aggresome",   
    25:  "Cytosol",   
    26:  "Cytoplasmic_bodies",   
    27:  "Rods_and_rings",
    28:  "All"
}

reverse_train_labels = dict((v,k) for k,v in label_names.items())

def fill_targets(row):
    try:
        row.Predicted = np.array(row.Predicted.split(" ")).astype(np.int)
    except:
        row.Predicted = []
    return row
```

```python
result0 = result0.apply(fill_targets, axis=1)
result0.head()
```

```python
result1 = pd.read_csv(os.path.join(result_dir, 'protein_classification_0.csv'))
#result1 = pd.read_csv(os.path.join(result_dir, 'submission-17.csv'))
result1.head()
```

```python
result1 = result1.apply(fill_targets, axis=1)
result1.head()
```

```python
pred_bing = []
for i, image_id in enumerate(result0['Id']):
    r0 = result0.Predicted[i]
    r1 = result1.Predicted[i]
    r_bing = np.concatenate([r0, r1]).astype(np.int)
    r_bing = np.unique(r_bing)
    r_bing = ' '.join(str(x) for x in r_bing)
    #r_jiao = ' '.join(str(x) for x in r_jiao)
    pred_bing.append({'Id': image_id, 'Predicted': r_bing})
    #pred_jiao.append({'Id': image_id, 'Predicted': r_jiao})
```

```python
submission_bing = pd.DataFrame(pred_bing)[['Id', 'Predicted']]
submission_bing.to_csv(os.path.join(result_dir, 'submission-bing-20-0.csv'), index=False)
```

```python
pred_jiao = []
for i, image_id in enumerate(result0['Id']):
    r0 = result0.Predicted[i]
    r1 = result1.Predicted[i]
    if len(r1) == 0:
        r_jiao = r0
    else:
        r1 = np.array(r1).astype(np.int)
        r_jiao = np.intersect1d(r0, r1).astype(np.int)
    if len(list(r_jiao)) == 0:
        r_jiao = r0
    r_jiao = ' '.join(str(x) for x in r_jiao)
    pred_jiao.append({'Id': image_id, 'Predicted': r_jiao})
submission_jiao = pd.DataFrame(pred_jiao)[['Id', 'Predicted']]
submission_jiao.to_csv(os.path.join(result_dir, 'submission-jiao-16-17.csv'), index=False)
```
