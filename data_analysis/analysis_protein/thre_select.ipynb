```python
import os
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
sns.set()
```

```python
pd_eval = pd.read_csv(os.path.join('../../result/protein', 'protein_eval.csv'))
pd_eval.head()
```

```python
label_names = {
    0:  "Nucleoplasm",  
    1:  "Nuclear_membrane",   
    2:  "Nucleoli",   
    3:  "Nucleoli_fibrillar_center",   
    4:  "Nuclear_speckles",
    5:  "Nuclear_bodies",   
    6:  "Endoplasmic_reticulum",   
    7:  "Golgi_apparatus",   
    8:  "Peroxisomes",   
    9:  "Endosomes",   
    10:  "Lysosomes",   
    11:  "Intermediate_filaments",   
    12:  "Actin_filaments",   
    13:  "Focal_adhesion_sites",   
    14:  "Microtubules",   
    15:  "Microtubule_ends",   
    16:  "Cytokinetic_bridge",   
    17:  "Mitotic_spindle",   
    18:  "Microtubule_organizing_center",   
    19:  "Centrosome",   
    20:  "Lipid_droplets",   
    21:  "Plasma_membrane",   
    22:  "Cell_junctions",   
    23:  "Mitochondria",   
    24:  "Aggresome",   
    25:  "Cytosol",   
    26:  "Cytoplasmic_bodies",   
    27:  "Rods_and_rings",
    28:  "All"
}
```

```python
def str_to_float(row):
    for label_name in label_names.values():
        row[label_name] = np.array(row[label_name].split(' ')).astype(np.float)
    row["Thre"] = int(row["Thre"])
    return row
```

```python
pd_eval = pd_eval.apply(str_to_float, axis=1)
pd_eval.head()
```

```python
plt.subplots(figsize=(15, 40))
x_thre = pd_eval['Thre'].values[1:]
for i in label_names:
    plt.subplot(10,3,i+1)
    y_eval = np.array(list(pd_eval[label_names[i]].values))[1:,0]
    plt.plot(x_thre, y_eval)
    plt.title(label_names[i])
    print(x_thre[np.argmax(y_eval)], np.max(y_eval))
plt.show()
```

```python
plt.subplots(figsize=(15, 40))
x_thre = pd_eval['Thre'].values[1:][::-1]
for i in label_names:
    plt.subplot(10,3,i+1)
    y_eval = np.array(list(pd_eval[label_names[i]].values))[1:,0][::-1]
    plt.plot(x_thre, y_eval)
    plt.title(label_names[i])
    print(x_thre[np.argmax(y_eval)], np.max(y_eval))
plt.show()
```
